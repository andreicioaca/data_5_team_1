```python
def batch_pickle(data, nb):
  batch_nb = nb
  batch_size = int( (len(data)/ batch_nb ))
  
  print(batch_size)
  start = 0

  for i in range(batch_nb -1) :
    end = start + batch_size
    print(start, start + batch_size)

    batch = data.iloc[start:end ,]
    batch.to_pickle(f"/content/drive/MyDrive/DevelHope/Google Playstore Data/batch{i}.pkl")

    start = end

  print(start, len(data))
  batch = data.iloc[start : len(data) , ]
  batch.to_pickle(f"/content/drive/MyDrive/DevelHope/Google Playstore Data/batch{nb - 1 }.pkl")
```

```python
def unpickle_merge(order):
  list = order
  
  first = list[0]
  data2 = pd.read_pickle(f"/content/drive/MyDrive/DevelHope/Google Playstore Data/batch{first-1}.pkl") 

  for i in range(1,len(list)):

    index = list[i]
    temp_data = pd.read_pickle(f"/content/drive/MyDrive/DevelHope/Google Playstore Data/batch{index-1}.pkl") 
    data2 = pd.concat( [data2, temp_data] )

  return data2
```

```python
def sliding_window(data,src_size, tgt_size, step):

  src_start = 0
  src_end = src_size

  result = list ()
  while src_start < len(data) - src_size + tgt_size :


    source = data.iloc[src_start: src_end, ]
    target = data.iloc[src_end: src_end + tgt_size , ]

    result.append( [source, target])

    print(f"source: {(src_start, src_end )}, target: {( src_end , src_end + tgt_size)}")

    src_start = src_start + step
    src_end = src_start + src_size


  return result
```
